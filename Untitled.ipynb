```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Constants
DATA_DIR = 'Data/AQI'
YEARS = range(2013, 2019)
DAYS_IN_YEAR = 365
CHUNKSIZE = 24  # Assuming hourly data for a day

# Create the directory if it doesn't exist
os.makedirs(DATA_DIR, exist_ok=True)

# Function to generate sample PM2.5 data and save to CSV
def generate_sample_data(year):
    # Generate random PM2.5 values (for example, between 0 and 150)
    pm25_data = np.random.uniform(0, 150, size=(DAYS_IN_YEAR, CHUNKSIZE))
    
    # Create a DataFrame
    df = pd.DataFrame(pm25_data, columns=[f'Hour_{i+1}' for i in range(CHUNKSIZE)])
    
    # Add a 'PM2.5' column that averages the hourly data
    df['PM2.5'] = df.mean(axis=1)
    
    # Save to CSV
    df.to_csv(f'{DATA_DIR}/aqi{year}.csv', index=False)

# Function to calculate average PM2.5 for a given year
def avg_data(year):
    average = []
    try:
        for rows in pd.read_csv(f'{DATA_DIR}/aqi{year}.csv', chunksize=CHUNKSIZE):
            add_var = 0
            count = 0
            for i in rows['PM2.5']:
                if isinstance(i, (float, int)):
                    add_var += i
                    count += 1
                elif isinstance(i, str) and i not in ['NoData', 'PwrFail', '---', 'InVld']:
                    add_var += float(i)
                    count += 1
            if count > 0:
                avg = add_var / count
                average.append(avg)
            else:
                average.append(0)  # Append 0 if no valid data
    except FileNotFoundError:
        print(f"File for year {year} not found.")
    return average

# Generate CSV files for each year
for year in YEARS:
    generate_sample_data(year)

# Calculate averages for each year
averages = {year: avg_data(year) for year in YEARS}

# Plotting the data
for year, data in averages.items():
    plt.plot(range(len(data)), data, label=f"{year} data")

plt.xlabel('Day')
plt.ylabel('PM 2.5')
plt.legend(loc='upper right')
plt.title('Average PM 2.5 Levels (2013-2018)')
plt.show()
```

```python
import os
import csv
import sys
import pandas as pd
from bs4 import BeautifulSoup

# Function to extract meteorological data from HTML
def met_data(month, year):
    # Read the HTML file
    file_path = f'Data/Html_Data/{year}/{month}.html'
    with open(file_path, 'rb') as file_html:
        plain_text = file_html.read()

    tempD = []
    finalD = []

    # Parse HTML content with BeautifulSoup
    soup = BeautifulSoup(plain_text, "lxml")
    for table in soup.findAll('table', {'class': 'medias mensuales numspan'}):
        for tbody in table:
            for tr in tbody:
                a = tr.get_text()
                tempD.append(a)

    # Calculate number of rows based on the data
    rows = len(tempD) // 15

    # Rearrange data into rows
    for _ in range(round(rows)):
        newtempD = [tempD.pop(0) for _ in range(15)]
        finalD.append(newtempD)

    # Clean up final data
    finalD.pop()  # Remove last row
    finalD.pop(0)  # Remove the first row

    # Remove unnecessary columns
    for row in finalD:
        del row[6]  # Remove column index 6
        del row[9:14]  # Remove columns index 9 to 13
        del row[0]  # Remove column index 0

    return finalD

# Function to combine data from CSV files
def data_combine(year, chunk_size):
    combined_data = []
    for chunk in pd.read_csv(f'Data/Real-Data/real_{year}.csv', chunksize=chunk_size):
        combined_data.extend(chunk.values.tolist())
    return combined_data

# Main execution block
if __name__ == "__main__":
    # Create directory for real data if it doesn't exist
    os.makedirs("Data/Real-Data", exist_ok=True)

    # Loop through each year and process data
    for year in range(2013, 2017):
        final_data = []
        # Create CSV file and write header
        with open(f'Data/Real-Data/real_{year}.csv', 'w', newline='') as csvfile:
            wr = csv.writer(csvfile)
            wr.writerow(['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM', 'PM 2.5'])

        # Loop through each month
        for month in range(1, 13):
            temp = met_data(month, year)
            final_data.extend(temp)

        # Get average PM2.5 data for the year
        pm = getattr(sys.modules[__name__], f'avg_data_{year}')()

        # Debugging output
        print(f"Year: {year}, PM Length: {len(pm)}, Final Data Length: {len(final_data)}")

        # Handle potential length mismatch
        if len(pm) < len(final_data):
            pm.extend(['-'] * (len(final_data) - len(pm)))  # Fill with '-' if pm is shorter

        # Insert PM2.5 data into final data
        for i in range(len(final_data)):
            final_data[i].insert(8, pm[i])  # Insert PM2.5 data at the 9th position

        # Write combined data to the CSV file
        with open(f'Data/Real-Data/real_{year}.csv', 'a', newline='') as csvfile:
            wr = csv.writer(csvfile)
            for row in final_data:
                if all(elem not in ["", "-"] for elem in row):
                    wr.writerow(row)

    # Combine data from all years into one CSV
    total_data = []
    for year in range(2013, 2017):
        total_data.extend(data_combine(year, 600))

    with open('Data/Real-Data/Real_Combine.csv', 'w', newline='') as csvfile:
        wr = csv.writer(csvfile)
        wr.writerow(['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM', 'PM 2.5'])
        wr.writerows(total_data)

    # Read the combined data into a DataFrame
    df = pd.read_csv('Data/Real-Data/Real_Combine.csv')
```
